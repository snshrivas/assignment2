#### 3. Implement a function longestWord() that takes a list of words and returns the longest one.


```python
def longestword(list): 
    max=0
    list2=[]
    for x in list:
        y=len(x)
        if max<y:
            max=y
    for y in list:
        if max==len(y):
            list2.append(y)
    return(list2)        
```

```python
i=1
list=[]
x=int(input("how many name do you want to enter:"))
print("please enter", x ,"name.")
while i<=x:
    y=input()
    list.append(y)
    i=i+1
print(list)    
```

```python
result=longestword(list)
for x in result:
    print(x)
```

```python
class A:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
        
class B(A):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)
        self.s=int((self.a+self.b+self.c)/2)
    def find_area(self):
        self.a=((self.s)-(self.a))*((self.s)-(self.b))*((self.s)-(self.c))
        self.b=(self.a*self.s)
        print(self.b**0.5)
        
        
  
```

#### 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(list,n):
    list1=[]
    for x in list:
        if len(x)>n:
            list1.append(x)
    return(list1)    
```

```python
i=1
list=[]
x=int(input("how many name do you want to enter:"))
print("please enter", x ,"name.")
while i<=x:
    y=input()
    list.append(y)
    i=i+1
print(list)
```

```python
n=int(input("Enter a number:"))
```

```python
result=filter_long_words(list,n)
print(result)
```

#### 2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words . Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4] Here 2,3 and 4 are the lengths of the words in the list.

```python
def map_word_to_length(list):
    length=[]
    for x in list:
        length.append(len(x))
    return(length)
```

```python
i=1
list=[]
x=int(input("how many name do you want to enter:"))
print("please enter", x ,"name.")
while i<=x:
    y=input()
    list.append(y)
    i=i+1
print(list)
```

```python
result=map_word_to_length(list)
print(result)
```

#### 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def check_vowel(ch):
    if ch in ['a','e','i','o','u','A','E','I','O','U']:
        return True
    else:
        return False
```
